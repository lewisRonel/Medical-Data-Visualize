#### ANALYSIS OF MEDICAL DATA

```python
#importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import data
df = pd.read_csv("medical_examination.csv")
df.head()
```

```python

df["overweight"] = (df["weight"] / ((df["height"]/100)**2) ).apply(lambda x: 1 if x >25 else 0)

df.head()
```

```python
# Normalize data by making 0 always good and 1 always bad. 
# If the value of 'cholesterol' or 'gluc' is 1, make the value 0. 
# If the value is more than 1, make the value 1.
df["cholesterol"] = df["cholesterol"].apply(lambda x:0 if x==1 else 1)
df["gluc"] = df["gluc"].apply(lambda x:0 if x==1 else 1)
```

```python
df["overweight"].value_counts()
```

```python
# Create DataFrame for cat plot using `pd.melt` using just the values from 'cholesterol', 'gluc', 'smoke', 'alco', 'active', and 'overweight'.
df_cat = pd.melt(df, id_vars=["cardio"], value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active','overweight'])


# Group and reformat the data to split it by 'cardio'. Show the counts of each feature. 
# You will have to rename one of the columns for the catplot to work correctly.
df_cat['total'] = 1
df_cat = df_cat.groupby(["cardio", "variable",'value'], as_index=False).count()
#draw with sns.catplot()

fig = sns.catplot(data=df_cat, x="variable", y="total", hue="value", kind="bar", col="cardio").fig
```

```python
# Clean the data

df_heat = df[
    (df['ap_lo'] <= df['ap_hi']) &
    (df['height'] >= df['height'].quantile(0.025)) &
    (df['height'] <= df['height'].quantile(0.975)) &
    (df['weight'] >= df['weight'].quantile(0.025)) &
    (df['weight'] <= df['weight'].quantile(0.975))
]
```

```python
# Calculate the correlation matrix
corr = df_heat.corr(method="pearson")
corr
```

```python
# Generate a mask for the upper triangle
mask = np.triu(corr)
```

```python
#set the matplotlib figure
fig, ax = plt.subplots(figsize = (12,12)) #create a figure containing a single axe
#draw the heatmap 
sns.heatmap(corr,  mask=mask, linewidths=1, annot = True,fmt=".1F", center= 0.08, cbar_kws={"shrink": 0.5})
```
